```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Mechademy-Git/DS-Intern-Assignment/main/data/data.csv")
df.head()
```

```python
print("Summary Statistics:")
df.describe()
```

```python
# Univariate Analysis
# Numerical summary statistics


# Create histograms for energy consumption
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['equipment_energy_consumption'], kde=True)
plt.title('Equipment Energy Consumption Distribution')
plt.subplot(1, 2, 2)
sns.histplot(df['lighting_energy'], kde=True)
plt.title('Lighting Energy Distribution')
plt.tight_layout()
plt.show()

# Temperature distributions across zones
plt.figure(figsize=(15, 10))
for i in range(1, 10):
    plt.subplot(3, 3, i)
    sns.histplot(df[f'zone{i}_temperature'], kde=True)
    plt.title(f'Zone {i} Temperature')
plt.tight_layout()
plt.show()

# Humidity distributions across zones
plt.figure(figsize=(15, 10))
for i in range(1, 10):
    plt.subplot(3, 3, i)
    sns.histplot(df[f'zone{i}_humidity'], kde=True)
    plt.title(f'Zone {i} Humidity')
plt.tight_layout()
plt.show()
```

Let's break down the analysis so far:

1. The summary statistics table provides a snapshot of the central tendency and spread for each variable, including temperatures, humidity, and energy consumption.
2. The first histogram shows the distribution of equipment and lighting energy consumption, which helps us understand their typical usage patterns and any outliers.
3. The next two sets of histograms visualize the temperature and humidity distributions across all zones, revealing how environmental conditions vary throughout the dataset.
Next, I'll move on to bivariate analysis to explore relationships between key variables.

```python
# Bivariate Analysis
# Correlation matrix for numerical variables
corr = df.corr(numeric_only=True)
plt.figure(figsize=(16, 10))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Scatter plot: Equipment energy vs. lighting energy
plt.figure(figsize=(7, 5))
sns.scatterplot(x='equipment_energy_consumption', y='lighting_energy', data=df)
plt.title('Equipment Energy vs. Lighting Energy')
plt.xlabel('Equipment Energy Consumption')
plt.ylabel('Lighting Energy')
plt.show()

# Scatter plot: Outdoor temperature vs. equipment energy consumption
plt.figure(figsize=(7, 5))
sns.scatterplot(x='outdoor_temperature', y='equipment_energy_consumption', data=df)
plt.title('Outdoor Temperature vs. Equipment Energy Consumption')
plt.xlabel('Outdoor Temperature')
plt.ylabel('Equipment Energy Consumption')
plt.show()
```

- The correlation matrix heatmap below shows how strongly each pair of numerical variables is related. Strong positive or negative values indicate a strong relationship.

- The scatter plot of equipment energy vs. lighting energy helps us see if higher equipment usage is associated with higher lighting usage.

- The scatter plot of outdoor temperature vs. equipment energy consumption helps us understand if energy use is affected by outside weather.



```python
# Multivariate Analysis
# Pairplot for selected variables to visualize interactions
selected_cols = ['equipment_energy_consumption', 'lighting_energy', 'outdoor_temperature', 'zone1_temperature', 'zone1_humidity']
sns.pairplot(df[selected_cols].dropna())
plt.suptitle('Pairplot of Key Variables', y=1.02)
plt.show()

# Multiple regression: Predict equipment energy consumption from other variables
import statsmodels.api as sm
X = df[['lighting_energy', 'outdoor_temperature', 'zone1_temperature', 'zone1_humidity']].dropna()
y = df.loc[X.index, 'equipment_energy_consumption']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Calculate average energy consumption and environmental metrics per zone
zone_metrics = pd.DataFrame()

# Get temperature and humidity averages for each zone
for i in range(1, 10):  # Zones 1-9
    temp_col = f'zone{i}_temperature'
    humid_col = f'zone{i}_humidity'
    
    if temp_col in df.columns and humid_col in df.columns:
        zone_metrics.loc[f'Zone {i}', 'Avg Temperature'] = pd.to_numeric(df[temp_col], errors='coerce').mean()
        zone_metrics.loc[f'Zone {i}', 'Avg Humidity'] = pd.to_numeric(df[humid_col], errors='coerce').mean()
        
# Calculate efficiency metrics
zone_metrics['Temperature Stability'] = zone_metrics['Avg Temperature'].std()
zone_metrics = zone_metrics.sort_values('Temperature Stability')

print("\
Zone Efficiency Metrics:")
print(zone_metrics)

# Visualize temperature distribution across zones
plt.figure(figsize=(12, 6))
temp_data = pd.DataFrame()
for i in range(1, 10):
    col = f'zone{i}_temperature'
    if col in df.columns:
        temp_data[f'Zone {i}'] = pd.to_numeric(df[col], errors='coerce')

sns.boxplot(data=temp_data)
plt.title('Temperature Distribution Across Zones')
plt.ylabel('Temperature')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyze energy consumption trends over time
# First, ensure timestamp is in datetime format

df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')

# Sort by timestamp
trend_df = df.sort_values('timestamp')

# Resample to daily averages for smoother trends
trend_df.set_index('timestamp', inplace=True)
daily_trends = trend_df[['equipment_energy_consumption', 'lighting_energy']].apply(pd.to_numeric, errors='coerce').resample('D').mean()

# Show head of daily trends
daily_trends_head = daily_trends.head()
print(daily_trends_head)

# Plot the trends
daily_trends.plot(figsize=(14,6))
plt.title('Daily Energy Consumption Trends')
plt.ylabel('Energy Consumption')
plt.xlabel('Date')
plt.legend(['Equipment Energy', 'Lighting Energy'])
plt.tight_layout()
plt.show()
```

Key findings from the time series analysis:

- Equipment energy shows higher variability than lighting energy
- Daily equipment energy ranges from ~70 to 150 units
- Lighting energy remains relatively stable around 3-27 units
- Clear cyclical patterns visible in both metrics

```python
# Step 1: Convert energy columns to numeric before grouping
energy_cols = ['equipment_energy_consumption', 'lighting_energy']
for col in energy_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Step 2: Group by hour and calculate mean
hourly_avg = df.groupby('hour')[energy_cols].mean()

# Step 3: Show the head of the hourly averages
print(hourly_avg.head())

# Step 4: Plot hourly energy consumption patterns
plt.figure(figsize=(12, 6))
hourly_avg.plot()
plt.title('Average Energy Consumption by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Energy Consumption')
plt.legend(['Equipment Energy', 'Lighting Energy'])
plt.grid(True)
plt.show()

# Step 5: Identify peak usage hours
peak_equipment = hourly_avg['equipment_energy_consumption'].nlargest(3)
peak_lighting = hourly_avg['lighting_energy'].nlargest(3)
print('\
Peak Equipment Energy Hours:')
print(peak_equipment)
print('\
Peak Lighting Energy Hours:')
print(peak_lighting)
```

Peak energy usage occurs at:

- Equipment: Highest between 17:00-19:00 (up to 195 units)
- Lighting: Peaks 20:00-21:00 (around 9.7 units)

```python
# Convert weekday numbers to names for better readability
df['weekday_name'] = df['timestamp'].dt.day_name()

# Calculate daily averages
daily_avg = df.groupby('weekday_name')[['equipment_energy_consumption', 'lighting_energy']].mean()

# Sort by weekday order
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_avg = daily_avg.reindex(weekday_order)

# Plot
plt.figure(figsize=(12, 6))
daily_avg.plot(kind='bar')
plt.title('Average Energy Consumption by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Energy Consumption')
plt.xticks(rotation=45)
plt.legend(['Equipment Energy', 'Lighting Energy'])
plt.tight_layout()
plt.show()

# Display averages
print("\
Daily Averages:")
print(daily_avg)
```

The chart above shows the average equipment and lighting energy consumption for each day of the week. You can see that equipment energy usage peaks on Mondays and is lowest midweek, while lighting energy is highest on Mondays and generally lower on Fridays and Saturdays. This suggests that the start of the week is the most energy-intensive, possibly due to operational patterns or increased activity after the weekend.


```python
# Create weekday/weekend flag
df['is_weekend'] = df['weekday_name'].isin(['Saturday', 'Sunday'])

# Calculate statistics for weekdays vs weekends
weekday_stats = df.groupby('is_weekend')[['equipment_energy_consumption', 'lighting_energy']].agg(['mean', 'std'])

# Plot comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Box plot for equipment energy
df.boxplot(column='equipment_energy_consumption', by='is_weekend', ax=ax1)
ax1.set_xticklabels(['Weekday', 'Weekend'])
ax1.set_title('Equipment Energy Distribution')

# Box plot for lighting energy
df.boxplot(column='lighting_energy', by='is_weekend', ax=ax2)
ax2.set_xticklabels(['Weekday', 'Weekend'])
ax2.set_title('Lighting Energy Distribution')

plt.tight_layout()
plt.show()

print("\
Weekday vs Weekend Statistics:")
print(weekday_stats)
```

Here’s the comparison of energy usage between weekdays and weekends:

- Equipment energy consumption is slightly higher on weekends, but the difference is not dramatic.
- Lighting energy usage is a bit lower on weekends compared to weekdays.
- The boxplots above show the distribution and variability for both metrics.
- This suggests that while there’s a small increase in equipment energy on weekends, lighting energy tends to drop, possibly due to reduced occupancy or different operational patterns. If you want to explore specific days, time periods, or look for anomalies, just let me know!
